### Exercise 1:
Create a simple line plot for `y = x` for `x` values ranging from 0 to 10.

````python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
x = np.arange(0,11)
y = x
plt.figure(figsize=(8,6))
plt.plot(x,y,'blue')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot x against y')
plt.show()
````

### Exercise 7:
Plot two line plots on the same graph: `y1 = x^2` and `y2 = x^3` for `x` ranging from 0 to 10.

````python
x = np.arange(0,11)
y1 = x**2
y2 = x**3
plt.figure(figsize = (10,8), dpi = 70)
plt.plot(x,y1, 'r')
plt.plot(x,y2, 'g')
# plt.xlabel('x')
# plt.ylabel('y')
# plt.title(r'Graph of $y_1 = x^2$ and $y_2 = x^3$ for $x$ ranging from 0 to 10')

fig, axes = plt.subplots(1, 2, figsize=(8,4), dpi=100)

# plot subplot 1
axes[0].plot(x, y1, color="green", label="y = x**2")
axes[0].plot(x, y2, color="red", label="y = x**3")
axes[0].legend(loc=2); # upper left corner
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title(r'Graph of $y_1 = x^2$ and $y_2 = x^3$ for $x$ ranging from 0 to 10')

# plot subplot 2
axes[1].plot(x, x**2, color="violet", label="y = x**2")
axes[1].plot(x, x**3, color="blue", label="y = x**3")
axes[1].legend(loc=2); # upper left corner
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].set_title('Plot of y=x^2 and y=x^3')

# `fig.tight_layout()` automatically adjusts the positions of the axes on the figure canvas so that there is no overlapping content
# comment this out to see the difference
fig.tight_layout()

plt.show()
````

### Exercise 17:
Create a plot with multiple subplots in a 2x2 grid. Plot `y1 = x`, `y2 = x^2`, `y3 = x^3`, and `y4 = x^4`.

````python
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
x=  np.arange(0,11,0.1)
y1 = x
y2 = x**2
y3 = x**3
y4 = x**4
fig, axes = plt.subplots(2, 2, figsize=(10,10), dpi=100)

# plot subplot 1
axes[0,0].plot(x, y1, color="green", label=r"$y = x$")
axes[0,0].legend(loc=2); # upper left corner
axes[0,0].set_xlabel('x')
axes[0,0].set_ylabel('y1')
axes[0,0].set_title(r'Plot for $y=x$')

# plot subplot 2
axes[1,0].plot(x, y2, color="blue", label=r"$y = x^2$")
axes[1,0].legend(loc=2); # upper left corner
axes[1,0].set_xlabel('x')
axes[1,0].set_ylabel('y2')
axes[1,0].set_title(r'Plot for $y=x^2$')

# plot subplot 3
axes[1,1].plot(x, y3, color="violet", label=r"$y = x^3$")
axes[1,1].legend(loc=2); # upper left corner
axes[1,1].set_xlabel('x')
axes[1,1].set_ylabel('y3')
axes[1,1].set_title(r'Plot for $y=x^2$')

# plot subplot 4
axes[0,1].plot(x, y4, color="red", label=r"$y = x^4$")
axes[0,1].legend(loc=2); # upper left corner
axes[0,1].set_xlabel('x')
axes[0,1].set_ylabel('y4')
axes[0,1].set_title(r'Plot for $y=x^3$')

fig.tight_layout()
plt.show()
````
